```python
# Function to split text into paragraphs with a maximum length of tokens
tokenizer = pipe.tokenizer

def split_into_paragraphs(text, max_length=512):
    paragraphs = []
    current_paragraph = ""

    # Tokenize the text
    tokens = tokenizer.tokenize(tokenizer.decode(tokenizer.encode(text)))

    for token in tokens:
        if len(tokenizer.encode(current_paragraph + token)) <= max_length:
            current_paragraph += token + " "
        else:
            paragraphs.append(current_paragraph.strip())
            current_paragraph = token + " "

    if current_paragraph:
        paragraphs.append(current_paragraph.strip())

    return paragraphs

# Example usage:
book_text =data
paragraphs = split_into_paragraphs(book_text)

```

```python
import pandas as pd
results_df = pd.DataFrame(columns=["LLM Output"])
data_superV=pd.DataFrame(columns=["text","LLM Output"])
```

```python
# start at 4:45
# 100 pargraphs take between 1 hour to 1 and half hour
# continue from index 50 pargraph[50]
for i in range(len(sub_pragraphs)):

  messages = [
    {
        "role": "system",
        "content": "Given the following text, clean and extract sentences suitable for fine-tuning a medical model,the output should be in form of one sentence to give to a model for fine tune",

    },
    {"role": "user", "content": f"{sub_pragraphs[i]}"},
  ]
  x = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
  output = pipe(x, max_new_tokens=512, do_sample=True, temperature=0.7, top_k=50, top_p=0.95)
  index = output[0]['generated_text'].find('<|assistant|>')
  output =output[0]['generated_text'][index+15:-1]

  results_df = results_df.append({ "LLM Output": output}, ignore_index=True)
  data_superV = data_superV.append({ "text":sub_pragraphs[i],"LLM Output": output}, ignore_index=True)

  print(len(data_superV))

```
